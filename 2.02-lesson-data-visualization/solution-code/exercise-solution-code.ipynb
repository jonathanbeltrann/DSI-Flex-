## Time to make some plots

**4. Plot the percentage of people that say soccer is their favorite sport over time. Give this plot a descriptive title and axis labels. Interpret this plot - what information/insight do you get from this plot?**

```python
# Size
plt.figure(figsize = (12, 7))

# Line chart
plt.plot(sports['Year'], sports['Soccer'])

# Create a descriptive title
plt.title('Soccer as a Favorite Sport Over Time')

# Add axis labels
plt.ylabel('Percentage of People'); 
# X label is unnecessary here since it is date, inferred
```

> **Interpretation:** Soccer has steadily increased as people's favorite sport from 2007 - 2017.

**4. Plot a bar chart of all of the sports and the percentage of people who say that was their favorite sport in 2017. Give this plot a descriptive title and axis labels. Interpret this plot - what information/insight do you get from this plot?**

*Hint:* You may need to access the column names here!

```python
# Size
plt.figure(figsize = (15, 5))

# Bar chart
# Horizontal so you can more easily read the names
plt.barh(sports.columns[1:], sports.loc[0,'Football':])

# Create a descriptive title
plt.title('Favorite Sports in 2017')

# Add axis labels
plt.xlabel('Percentage of People'); 
```

> **Interpretation:** Football has the highest percentage of people claiming the sport as their favorite. Bowling, wrestling, swimming, fishing, track and field, and rodeo do not have any supporters.

**Bonus**: Sort the above plot so that the bars are in descending order (highest percentage/largest bar to lowest percentage/smallest bar). Make sure the labels match up!

```python
# Bonus: sort this from highest to lowest

# save data as new dataframe 
sports_2017 = pd.DataFrame({'sport': sports.columns[1:], 'percentage': sports.loc[0,'Football':]})

# sort values by percent
sports_2017.sort_values(by = 'percentage', inplace = True)

sports_2017.head()
```

```python
# Size
plt.figure(figsize = (15, 5))

# Bar chart
# Horizontal so you can more easily read the names
plt.barh(sports_2017['sport'], sports_2017['percentage'])

# Create a descriptive title
plt.title('Favorite Sports in 2017')

# Add axis labels
plt.xlabel('Percentage of People'); 
```

**5. Advanced: Graph the popularity over time of two different sports on the same plot. Give this plot a descriptive title and axis labels. Make this plot look as visually appealing as possible. Interpret this plot - what information/insight do you get from this plot?**

```python
# Establish the size of the figure.
plt.figure(figsize = (16, 9))

# Create line plot of Soccer popularity over time.
plt.plot(sports['Year'], 
         sports['Soccer'],
         c = 'blue',
         label = 'Soccer')

# Create line plot of Ice Hockey popularity over time.
plt.plot(sports['Year'], 
         sports['Ice hockey'],
         c = 'orange',
         label = 'Hockey')

# Create the legend and set it in the upper-left hand corner.
plt.legend(loc = 'upper left', fontsize = 20)

# Create a very light grid in the background.
plt.grid(True, linewidth = 0.5, linestyle = '-', c = 'black', alpha = 0.1)

# Create larger tick mark labels on the Y axis
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('%', fontsize = 20, rotation = 0)

# Create title.
plt.title("Soccer has been a more popular sport to watch than Ice Hockey since 2013", fontsize = 23)

# Add vertical line (vlines) to specifically identify the point of interest.
plt.axvline(x = 2013, color = 'darkgrey');
```

> **Interpretation:** Since this plot is so clearly labeled, the interpretation is intuitive: soccer has been a more popular sport than ice hockey since 2013.
